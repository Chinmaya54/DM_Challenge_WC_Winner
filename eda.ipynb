Stats of Indian team


```python
# Sort the'results' DataFrame so that the rows where India participated as either team1 or team2 are included.
df = results[(results['team1'] == 'India') | (results['team2'] == 'India')]

# Make a new DataFrame called "India" that has every row from the DataFrame that has been filtered.
India = df.iloc[:]

# Show the 'India' DataFrame's initial few rows.
India.head()
```

```python
# Creates a new dataframe 'India_win' containing rows where the 'Winner' column is 'India'.
India_win = India[India['winner']=='India']
```

No.of wins in ODIs against other teams


```python
# Exclude Team India's name
excluded_value = 'India'

# Filtering out rows with the excluded value
filtered_df = India_win[India_win['team2'] != excluded_value]

# In the filtered DataFrame, count how many times each value appears.
value_counts = filtered_df['team2'].value_counts()

# Print the value counts
print(value_counts)
```

Displaying Team India's wins against other teams using bar graph

Win Percentage of India Against Each Team

```python
# Number of wins against each team
# These are the teams India has played in ODI matches against
team_win_counts = {
    'Australia': 54,
    'New Zealand': 58,
    'South Africa ': 37,
    'Pakistan': 	55,
    'Sri Lanka': 93,
    'Bangladesh': 30,
    'England': 57,
    'Netherlands': 2,
    'Afghanistan': 3
}

# Total matches played is calculated
total_matches = sum(team_win_counts.values())

# India's win percentages against each team is calculated
win_percentages = {team: (wins / total_matches) * 100 for team, wins in team_win_counts.items()}

# Pie chart
plt.figure(figsize=(5, 5))
plt.pie(win_percentages.values(), labels=win_percentages.keys(), autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Title for the pie chart
plt.title('Win Percentage of India Against Each Team')

# Display the pie chart
plt.show()
```
