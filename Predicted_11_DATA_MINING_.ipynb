```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
data=pd.read_csv("/content/batting_player_stats.csv")
indian_players = pd.DataFrame(data)
indian_players = data[data['Country'] == 'IND']

features = indian_players.drop(['Player'], axis=1)

indian_players['Highest Score'] = pd.to_numeric(indian_players['Highest Score'].astype(str).str.replace('*', ''), errors='coerce')

columns_to_check = ['Runs', 'Highest Score']
indian_players = indian_players.dropna(subset=columns_to_check)

X_train, X_test, y_train, y_test = train_test_split(features, indian_players['Runs'], test_size=0.2, random_state=42)

dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

dt_predictions = dt_model.predict(features)  # Use the entire dataset for predictions

best_batsmen_indices = dt_predictions.argsort()[-7:][::-1]
best_batsmen = indian_players.iloc[best_batsmen_indices]['Player'].tolist()

print("Best 7 Batsmen:")
print(best_batsmen)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(features, indian_players['Runs'], test_size=0.2, random_state=42)
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)

mae_dt = mean_absolute_error(y_test, dt_predictions)
mse_dt = mean_squared_error(y_test, dt_predictions)
r2_dt = r2_score(y_test, dt_predictions)

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

mae_rf = mean_absolute_error(y_test, rf_predictions)
mse_rf = mean_squared_error(y_test, rf_predictions)
r2_rf = r2_score(y_test, rf_predictions)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)

mae_lr = mean_absolute_error(y_test, lr_predictions)
mse_lr = mean_squared_error(y_test, lr_predictions)
r2_lr = r2_score(y_test, lr_predictions)

print("Decision Tree Metrics:")
print("MAE: {:.2f}".format(mae_dt))
print("MSE: {:.2f}".format(mse_dt))
print("R-squared: {:.2f}".format(r2_dt))
print("\n")

print("Random Forest Metrics:")
print("MAE: {:.2f}".format(mae_rf))
print("MSE: {:.2f}".format(mse_rf))
print("R-squared: {:.2f}".format(r2_rf))

print("Linear Regression Metrics:")
print("MAE: {:.2f}".format(mae_lr))
print("MSE: {:.2f}".format(mse_lr))
print("R-squared: {:.2f}".format(r2_lr))
```

```python
y = [0.87, 0.81, 1]
x = ['Decision Tree Metrics', 'Random Forest Metrics', 'Linear Regression Metrics']
plt.figure(figsize=(10, 6))
plt.bar(x, y, color=['blue', 'green', 'orange'])
plt.xlabel('Machine Learning Algorithms')
plt.ylabel('value')
plt.title(' R-squared Comparison of ML Algorithms')
plt.ylim(0, 1)
plt.show()
```

```python
columns_to_replace = ['Wkts', 'Ave', 'Econ', 'SR']
data[columns_to_replace] = data[columns_to_replace].replace(to_replace='[^0-9.]', value='0', regex=True)

numeric_columns = ['Wkts', 'Ave', 'Econ', 'SR']
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
ind_bowlers = data[data['Country'] == 'IND']
features_ind = ind_bowlers.drop(['Player', 'Country','Span','BBI'], axis=1)

columns_to_check_ind = ['Wkts', 'Ave', 'Econ', 'SR']
ind_bowlers = ind_bowlers.dropna(subset=columns_to_check_ind)

features_ind = features_ind.iloc[:len(ind_bowlers)]
y_ind = ind_bowlers['Wkts']

X_train_ind, X_test_ind, y_train_ind, y_test_ind = train_test_split(features_ind, y_ind, test_size=0.2, random_state=42)
X_train_ind.replace("-", 0, inplace=True)
y_train_ind.replace("-", 0, inplace=True)
features_ind.replace("-", 0, inplace=True)

dt_model_ind = DecisionTreeRegressor()
dt_model_ind.fit(X_train_ind, y_train_ind)
dt_predictions_ind = dt_model_ind.predict(features_ind)

rf_model_ind = RandomForestRegressor()
rf_model_ind.fit(X_train_ind, y_train_ind)
rf_predictions_ind = rf_model_ind.predict(features_ind)

lr_model_ind = LinearRegression()
lr_model_ind.fit(X_train_ind, y_train_ind)
lr_predictions_ind = lr_model_ind.predict(features_ind)
def evaluate_model(predictions, y_true):
  mae = mean_absolute_error(y_true, predictions)
  mse = mean_squared_error(y_true, predictions)
  r2 = r2_score(y_true, predictions)
  return mae, mse, r2

dt_mae_ind, dt_mse_ind, dt_r2_ind = evaluate_model(dt_predictions_ind, y_ind)
print("Decision Tree Metrics for IND:")
print(f"Mean Absolute Error: {dt_mae_ind}")
print(f"Mean Squared Error: {dt_mse_ind}")
print(f"R-squared: {dt_r2_ind}")

rf_mae_ind, rf_mse_ind, rf_r2_ind = evaluate_model(rf_predictions_ind, y_ind)
print("\nRandom Forest Metrics for IND:")
print(f"Mean Absolute Error: {rf_mae_ind}")
print(f"Mean Squared Error: {rf_mse_ind}")
print(f"R-squared: {rf_r2_ind}")

lr_mae_ind, lr_mse_ind, lr_r2_ind = evaluate_model(lr_predictions_ind, y_ind)
print("\nLinear Regression Metrics for IND:")
print(f"Mean Absolute Error: {lr_mae_ind}")
print(f"Mean Squared Error: {lr_mse_ind}")
print(f"R-squared: {lr_r2_ind}")

best_bowlers_dt_ind = ind_bowlers.iloc[dt_predictions_ind.argsort()[-4:][::-1]]['Player'].tolist()
best_bowlers_rf_ind = ind_bowlers.iloc[rf_predictions_ind.argsort()[-4:][::-1]]['Player'].tolist()
best_bowlers_lr_ind = ind_bowlers.iloc[lr_predictions_ind.argsort()[-4:][::-1]]['Player'].tolist()

print("\nBest 4 Bowlers for India - Decision Tree:")
print(best_bowlers_dt_ind)

print("\nBest 4 Bowlers for India - Random Forest:")
print(best_bowlers_rf_ind)

print("\nBest 4 Bowlers for India - Linear Regression:")
print(best_bowlers_lr_ind)
print("No data available for Indian bowlers.")
```
