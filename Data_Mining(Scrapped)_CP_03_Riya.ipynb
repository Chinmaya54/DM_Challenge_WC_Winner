#MOST RUNS

**Player:** The name of the cricket player who scored the runs.

**Country:** The country for which the player represents or plays international cricket.

**Span:** The time period (range of years) during which the player's career spanned.

**Matches:** The total number of matches the player participated in.

**Innings:** The total number of times the player has gone to bat in all matches.

**Not Outs:** The number of times the player remained not out after completing an inning.

**Runs:** The total number of runs scored by the player.

**Highest Score:** The highest individual score (number of runs) achieved by the player in a single inning.

**Average:** The average number of runs scored by the player per dismissal (batting average).

**Balls Faced:** The total number of balls faced by the player while batting.

**Strike Rate:** The strike rate of the player, calculated as (Runs/Balls Faced) * 100.

**Hundreds:** The number of times the player scored a century (100 runs or more in a single inning).

**Fifties:** The number of times the player scored a half-century (50 to 99 runs in a single inning).

**Zeros:** The number of times the player scored zero runs in an inning (got out without scoring).

**Fours:** The total number of boundaries (four runs) scored by the player.

**Sixes:** The total number of sixes (six runs) scored by the player.

###ANN

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Feature Selection
features = ['Innings', 'Average', 'Balls Faced', 'Hundreds', 'Fifties', 'Fours', 'Sixes', 'Batting_Average']
X = runs_df[features]
y = runs_df['Runs']

# Train-Test Split for the ANN model
X_train_ann, X_test_ann, y_train_ann, y_test_ann = train_test_split(X, y, test_size=0.2, random_state=52)

# Feature Scaling for ANN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_ann)
X_test_scaled = scaler.transform(X_test_ann)
```

```python
# Define function to create ANN model
def create_ann_model(learning_rate=0.001, batch_size=32):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1, activation='linear')
    ])

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Create and train ANN models with different batch sizes and learning rates
batch_sizes = [32, 64, 128]
learning_rates = [0.001, 0.01, 0.1]

best_model = None
best_mse = float('inf')

for batch_size in batch_sizes:
    for learning_rate in learning_rates:
        print(f"\nTraining ANN with Batch Size: {batch_size}, Learning Rate: {learning_rate}")
        ann_model = create_ann_model(learning_rate=learning_rate, batch_size=batch_size)
        ann_model.fit(X_train_scaled, y_train_ann, epochs=50, batch_size=batch_size)

        # Model Prediction
        y_pred_ann = ann_model.predict(X_test_scaled).flatten()

        # Model Evaluation
        mse_ann = mean_squared_error(y_test_ann, y_pred_ann)
        mae_ann = mean_absolute_error(y_test_ann, y_pred_ann)
        r2_ann = r2_score(y_test_ann, y_pred_ann)

        print(f'Mean Squared Error (ANN): {mse_ann}')
        print(f'Mean Absolute Error (ANN): {mae_ann}')
        print(f'R-squared (ANN): {r2_ann}')

        # Check if the current model is the best
        if mse_ann < best_mse:
            best_mse = mse_ann
            best_model = ann_model
```

```python
# Plotting Actual and Predicted Runs for the best ANN model
plt.figure(figsize=(10, 6))
plt.plot(y_test_ann.values, label='Actual Runs (ANN)', color='blue', marker='o')
plt.plot(best_model.predict(X_test_scaled).flatten(), label='Predicted Runs (ANN)', color='purple', marker='o')
plt.title('Actual vs Predicted Runs (ANN)')
plt.xlabel('Index')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
# Predicting runs for all batsmen in the training set (best ANN model)
batsman_data_ann = scaler.transform(X_train_ann)
predicted_runs_ann = best_model.predict(batsman_data_ann).flatten()

# Find the batsman with the highest predicted runs (best ANN model)
max_runs_index_ann = predicted_runs_ann.argmax()
batsman_with_max_runs_ann = runs_df.iloc[max_runs_index_ann]['Player']
country_of_max_runs_ann = runs_df.iloc[max_runs_index_ann]['Country']

print(f'\nBatsman with the most predicted runs (ANN): {batsman_with_max_runs_ann}')
print(f'Country of the batsman (ANN): {country_of_max_runs_ann}')
```

#MOST WICKETS

**Player:** The name of the cricket player who took the wickets.

**Country:** The country for which the player represents or plays international cricket.

**Span:** The time period (range of years) during which the player's career spanned.

**Matches**: The total number of matches the player participated in.

**Innings:** The total number of bowling innings bowled by the player.

**Balls:** The total number of balls bowled by the player.

**Overs:** The total number of overs bowled by the player. An over consists of six legal deliveries.

**MDNS:** The total number of maidens bowled by the player. A maiden over is an over in which no runs are scored.

**Runs:** The total number of runs conceded by the player.

**Wkts:** The total number of wickets taken by the player.

**BBI (Best Bowling in an Inning):** The player's best bowling performance in a single inning, represented as the number of wickets taken for the fewest runs.

**AVE (Bowling Average)**: The average number of runs conceded per wicket taken by the player.

**Econ (Economy Rate):** The average number of runs conceded per over bowled by the player.

**SR (Strike Rate):** The average number of balls bowled per wicket taken by the player.

**Fours:** The total number of times the player's deliveries were hit for four runs by the batsmen.

**Fives:** The total number of times the player took five wickets in a single inning.

###ANN

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Feature Selection
features = ['Matches', 'Innings', 'Balls', 'Overs', 'Runs', 'AVE', 'Econ', 'SR']
X = wickets_df[features]
y = wickets_df['Wkts']

# Train-Test Split for the ANN model
X_train_ann, X_test_ann, y_train_ann, y_test_ann = train_test_split(X, y, test_size=0.2, random_state=55)
# Feature Scaling for ANN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_ann)
X_test_scaled = scaler.transform(X_test_ann)
```

```python
# Define function to create ANN model
def create_ann_model(learning_rate=0.001, batch_size=32):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1, activation='linear')
    ])

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Create and train ANN models with different batch sizes and learning rates
batch_sizes = [32, 64, 128]
learning_rates = [0.001, 0.01, 0.1]

best_model = None
best_mse = float('inf')

for batch_size in batch_sizes:
    for learning_rate in learning_rates:
        print(f"\nTraining ANN with Batch Size: {batch_size}, Learning Rate: {learning_rate}")
        ann_model = create_ann_model(learning_rate=learning_rate, batch_size=batch_size)
        ann_model.fit(X_train_scaled, y_train_ann, epochs=50, batch_size=batch_size)

        # Model Prediction
        y_pred_ann = ann_model.predict(X_test_scaled).flatten()

        # Model Evaluation
        mse_ann = mean_squared_error(y_test_ann, y_pred_ann)
        mae_ann = mean_absolute_error(y_test_ann, y_pred_ann)
        r2_ann = r2_score(y_test_ann, y_pred_ann)

        print(f'Mean Squared Error (ANN): {mse_ann}')
        print(f'Mean Absolute Error (ANN): {mae_ann}')
        print(f'R-squared (ANN): {r2_ann}')

        # Check if the current model is the best
        if mse_ann < best_mse:
            best_mse = mse_ann
            best_model = ann_model
```

```python
# Plotting Actual and Predicted Wickets for the best ANN model
plt.figure(figsize=(10, 6))
plt.plot(y_test_ann.values, label='Actual Wickets (ANN)', color='blue', marker='o')
plt.plot(best_model.predict(X_test_scaled).flatten(), label='Predicted Wickets (ANN)', color='purple', marker='o')
plt.title('Actual vs Predicted Wickets (ANN)')
plt.xlabel('Index')
plt.ylabel('Wickets')
plt.legend()
plt.show()
```

```python
# Predicting wickets for all bowlers in the training set using the best ANN model
bowlers_data_ann = scaler.transform(X_train_ann)
predicted_wickets_ann = best_model.predict(bowlers_data_ann).flatten()

# Find the bowler with the most predicted wickets using the best ANN model
max_wickets_index_ann = predicted_wickets_ann.argmax()
bowler_with_max_wickets_ann = wickets_df.iloc[max_wickets_index_ann]['Player']
country_of_max_wickets_ann = wickets_df.iloc[max_wickets_index_ann]['Country']

print(f'\nBowler with the most predicted wickets (ANN): {bowler_with_max_wickets_ann}')
print(f'Country of the bowler (ANN): {country_of_max_wickets_ann}')
```
