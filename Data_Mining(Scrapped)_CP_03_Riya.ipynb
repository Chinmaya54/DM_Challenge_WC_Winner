###LINEAR REGRESSION

```python
runs_df.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Feature Selection
features = ['Innings', 'Average', 'Balls Faced', 'Hundreds', 'Fifties', 'Fours', 'Sixes', 'Batting_Average']
X = runs_df[features]
y = runs_df['Runs']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

# Model Training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Model Prediction
y_pred = model.predict(X_test)

# Model Evaluation
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')
```

```python
# Plotting Actual and Predicted Runs
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Actual Run (Linear Regression)', color='blue', marker='o')  # Actual Runs
plt.plot(y_pred, label='Predicted Runs (Linear Regression)', color='orange', marker='o')  # Predicted Runs
plt.title('Actual vs Predicted Runs')
plt.xlabel('Index')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
# Predicting runs for all batsmen in the training set
batsman_data = X_train
predicted_runs = model.predict(batsman_data)

# Find the batsman with the highest predicted runs
max_runs_index = predicted_runs.argmax()
batsman_with_max_runs = runs_df.iloc[max_runs_index]['Player']
country_of_max_runs = runs_df.iloc[max_runs_index]['Country']

print(f'Batsman with the most predicted runs: {batsman_with_max_runs}')
print(f'Country of the batsman: {country_of_max_runs}')
```

###RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor

# Train-Test Split for the second model
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X, y, test_size=0.2, random_state=55)

# Model Training (Random Forest)
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(X_train_rf, y_train_rf)

# Model Prediction (Random Forest)
y_pred_rf = model_rf.predict(X_test_rf)
```

```python
# Model Evaluation (Random Forest)
mse_rf = mean_squared_error(y_test_rf, y_pred_rf)
mae_rf = mean_absolute_error(y_test_rf, y_pred_rf)
r2_rf = r2_score(y_test_rf, y_pred_rf)

print(f'Mean Squared Error (Random Forest): {mse_rf}')
print(f'Mean Absolute Error (Random Forest): {mae_rf}')
print(f'R-squared (Random Forest): {r2_rf}')
```

```python
# Plotting Actual and Predicted Runs for Random Forest
plt.figure(figsize=(10, 6))
plt.plot(y_test_rf.values, label='Actual Runs (Random Forest)', color='blue', marker='o')
plt.plot(y_pred_rf, label='Predicted Runs (Random Forest)', color='green', marker='o')
plt.title('Actual vs Predicted Runs (Random Forest)')
plt.xlabel('Index')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
# Predicting runs for all batsmen in the training set (Random Forest)
batsman_data_rf = X_train_rf
predicted_runs_rf = model_rf.predict(batsman_data_rf)

# Find the batsman with the highest predicted runs (Random Forest)
max_runs_index_rf = predicted_runs_rf.argmax()
batsman_with_max_runs_rf = runs_df.iloc[max_runs_index_rf]['Player']
country_of_max_runs_rf = runs_df.iloc[max_runs_index_rf]['Country']

print(f'Batsman with the most predicted runs (Random Forest): {batsman_with_max_runs_rf}')
print(f'Country of the batsman (Random Forest): {country_of_max_runs_rf}')
```

###DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor

# Train-Test Split for the Decision Tree model
X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X, y, test_size=0.2, random_state=25)

# Model Training (Decision Tree)
model_dt = DecisionTreeRegressor(random_state=25)
model_dt.fit(X_train_dt, y_train_dt)
```

```python
# Model Prediction (Decision Tree)
y_pred_dt = model_dt.predict(X_test_dt)

# Model Evaluation (Decision Tree)
mse_dt = mean_squared_error(y_test_dt, y_pred_dt)
mae_dt = mean_absolute_error(y_test_dt, y_pred_dt)
r2_dt = r2_score(y_test_dt, y_pred_dt)

print(f'Mean Squared Error (Decision Tree): {mse_dt}')
print(f'Mean Absolute Error (Decision Tree): {mae_dt}')
print(f'R-squared (Decision Tree): {r2_dt}')
```

```python
# Plotting Actual and Predicted Runs for Decision Tree
plt.figure(figsize=(10, 6))
plt.plot(y_test_dt.values, label='Actual Runs (Decision Tree)', color='blue', marker='o')
plt.plot(y_pred_dt, label='Predicted Runs (Decision Tree)', color='red', marker='o')
plt.title('Actual vs Predicted Runs (Decision Tree)')
plt.xlabel('Index')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
# Predicting runs for all batsmen in the training set (Decision Tree)
batsman_data_dt = X_train_dt
predicted_runs_dt = model_dt.predict(batsman_data_dt)

# Find the batsman with the highest predicted runs (Decision Tree)
max_runs_index_dt = predicted_runs_dt.argmax()
batsman_with_max_runs_dt = runs_df.iloc[max_runs_index_dt]['Player']
country_of_max_runs_dt = runs_df.iloc[max_runs_index_dt]['Country']

print(f'Batsman with the most predicted runs (Decision Tree): {batsman_with_max_runs_dt}')
print(f'Country of the batsman (Decision Tree): {country_of_max_runs_dt}')
```
